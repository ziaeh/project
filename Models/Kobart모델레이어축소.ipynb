```python
from torch.utils.data import Dataset

class MyDataset(Dataset):
    def __init__(self, data):
        super().__init__()
        self.data = data
        self.input, self.target = self.tokenizing()

    def tokenizing(self):

      # max_len = max([len(passage) for passage in self.data['passage']])

      input_tokenized = []
      target_tokenized = []

      for i in range(len(self.data)):

          input_text = self.data.iloc[i, 0] #passage
          target_text = self.data.iloc[i, 1] #summary

          input_ids = tokenizer('<s>'+input_text+'</s>', padding="max_length", max_length = 1024, truncation=True)["input_ids"]
          target_ids = tokenizer('<s>'+target_text+'</s>', padding="max_length",max_length = 1024, truncation=True)["input_ids"]

          input_tokenized.append(input_ids)
          target_tokenized.append(target_ids)

      return torch.LongTensor(input_tokenized), torch.LongTensor(target_tokenized)

    def __len__(self):
        return len(self.input)

    def __getitem__(self, idx):
        return {
            'input_ids': self.input[idx],
            'labels': self.target[idx]
        }
```

```python
from sklearn.model_selection import train_test_split

dataset = MyDataset(data= df)

# 학습을 위해 train, test set으로 나눈다.
train_dataset, valid_dataset = train_test_split(dataset, test_size=0.1, random_state=42)
print(len(train_dataset), len(valid_dataset))
```

```python
from torch.utils.data import DataLoader

train_dataloader = DataLoader(train_dataset, batch_size= 20, shuffle = False)
valid_dataloader = DataLoader(valid_dataset, batch_size = 20, shuffle = False)
```

```python
import torch.nn as nn


epochs = 10
loss = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(params=model.parameters(), lr=0.001)
# scheduler =

for epoch in range(epochs):
    epoch_loss = 0.0
    for batch in train_dataloader:
        input = batch['input_ids'].to(device)
        target = batch['labels'].to(device)

        optimizer.zero_grad()

        pred = model(input, labels = target) #모르게떠

        # target = target.transpose(0, 1) #input, batch, embedding

        # 배치 전체에 대한 손실 계산
        batch_loss = pred.loss

        epoch_loss += batch_loss.item()

        batch_loss.backward()
        optimizer.step()

    epoch_loss /= len(train_dataloader)  # 에포크 평균 손실 계산
    print('Epoch:', '%02d' % (epoch + 1), 'cost =', '{:.6f}'.format(epoch_loss))

    # 그라디언트 초기화
    model.zero_grad()
```

```python
import torch.nn as nn

class MyModel(nn.Module):
  def __init__(self,pretrained_model, device, summarization_head_dropout=0.1):
    super().__init__()
    self.device = device
    self.embedding = pretrained_model
    self.fc = nn.Linear(30000, 768)
    self.activation = nn.GELU()
    self.dropout = nn.Dropout(summarization_head_dropout)
    self.summarization_head = nn.Linear(pretrained_model.config.d_model, 30000)

    for param in pretrained_model.get_encoder().parameters():
      param.requires_grad = False

  def forward(self, input, target):
    embed_output = self.embedding(input, labels = target)
    fc_output = self.fc(embed_output.logits) #모델 출력의 로짓(확률 분포를 나타내는 부분)을 포함하는 Tensor
    self.activation(fc_output)
    dropout_output = self.dropout(fc_output)
    output = self.summarization_head(dropout_output)

    return output
```

```python
new_model = MyModel(model, device)
```
